```
def get_point(point1, point2, len_from_p1=0.5):
    """
    point1, point2 - points coordinates as tuples
    """
    x_coord = point1[0] + (point2[0] - point1[0]) * len_from_p1
    y_coord = point1[1] + (point2[1] - point1[1]) * len_from_p1

    return [x_coord, y_coord, 0]
```

```
def add_texture(path_to_new_img, texture):
    img_m = io.imread(path_to_new_img)
    img_m = img_m[:,:,:-1]
    img_m_sum = img_m.sum(axis=2)
    mask = np.where(img_m_sum<1, 1, 0)
    mask_rgb = np.repeat(mask[:, :, np.newaxis], 3, axis=2)
    texture_resized = resize(texture, (img_m.shape[0] , img_m.shape[1]), anti_aliasing=True)
    texture_resized = texture_resized * 255
    texture_resized = texture_resized.astype('uint8')
    textured = texture_resized * mask_rgb.astype('uint8')
    out = img_m + textured
    #get which folder to use
    fname = path_to_new_img.split('/')[-1][:-10]
    if path_to_new_img[-9] == 'b':
        folder=1
        io.imsave('/content/drive/My Drive/Skoltech/Deep Learning/Dataset_for_project/{}/{}_btmask.png'.format(folder, fname), out)
    else:
        folder=2
        io.imsave('/content/drive/My Drive/Skoltech/Deep Learning/Dataset_for_project/{}/{}_gtmask.png'.format(folder, fname), out)
    
```
